# Arithmetic Operators

```python
print(5 * 5)
print(5 / 5)        # always float
print(5.22 / 5)     # always float
print(5.22 // 5)    # float
print(10 // 5)      # int
print(5. + 5)
print(5 + 5)
print(15 - 5)
print(15. - 5)
print(152. % -5)    # keeps the divisor's sign
print(2**8)
```

```python
print("123" * 2)
print("123" + "abc")
print([1,2,3] * 3)
print([1,2,3] + [1,1,2])
print([1,2,3].__mul__(3) )
print((1,2,3) * 3)
print((1,2,3) + (4,5,6))
#print({1,2,3} * 3)          # doesn't make sense to add numbers to the UNIQUE SET
```

# Conditional Operators

```python
print(5 > 5)
print(5 >= 5)
print(5 == 5)
print(5 != 5)
print(5 != "4jkfds")        # false
print(5 > 10 < 20)
print(5 < 10 < 20)          # 5 < 10 and 10 < 20
print([1,2,3] == [1,2,3])   # True, because checks are they the same
print([1,2,3] is [1,2,3])   # False, the same identity

ls = [1,2,3]
print(ls is ls)   # True, the same identity
print(ls is not ls)   # False, not the same identity
```
