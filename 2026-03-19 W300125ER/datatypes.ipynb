## Type Checking Functions

the functions: 
- `type(expression)`  check the type of the value. 
- `isinstance(expression, TYPE)` check the instance of a value is the same as class

```python
print(type(555))
print(type(555.5))
print(type("hello"))
print(type([5, 5, 6]))
print(type((1, 2, 3)))
print(type({5, 5, 5}))
print(type(frozenset([5, 5, 5])))
print(type({"a": 5, "b": 688}))
print(type(True))
print(type(None))
print(type(str))
print(type(dict))
print(type(tuple))
print(type(list))
print(type(set))
```

```python
print(isinstance("sdflkjsdf", str))
print(isinstance("sdflkjsdf", int))
```

## Data Types Syntax

### Strings (str)

```python
print('hello')             # single quote
print("hello")             # double quote
print("\" hello \n \\")    # using escape characters 

print("'")
print('"')

# Triple quotes - multiline (can be double or single quotes)
print("""hello
world""")

print('''hello
world''')

# continue in the next line. treated as a one line string (no break characters)
print("hello \
    world")


# Raw string - disables escape characters
print('\\')         # escapes the \ character and shows one
print(r'\\')        # does not escape any characters

"\n"   # newline
"\t"   # tab
"\\"   # backslash
"\""   # quote


```

#### Format Strings

```python
username = "pink_elephant"

# MODERN 
print(f"hello {1+1} {username}")        # concatenate expressions as part of the string
print(f"hello {{username}}")            # escape concatenation syntax 

my_string= "this {} can be {} used {username}"
print(my_string.format(1,2, username=username))       # in this case we fill the `{}` later and names are parameters
print(my_string.format(5, 7, username=username))       # in this case we fill the `{}` later and names are parameters


# LEGACY (old)
print("Hello %s" % "world")
print("%.2f" % 3.14159)

print(f"user: {username}")
```

#### Concatenation

```python
# implicit - can only be used with literal strings (no variables)
print("hello" "world")
text = (                # can be in different lines (as long as in `()`)
    "hello" 
    " " 
    "world"
)
print(text)


# explicit
print("hello, " + username) # with variable
```

### Int (int)

```python
# Defaults to base 10
42
-7
0

0b1010     # binary (base 2) → 10
0o12       # octal  (base 8) → 10
0xA        # hex    (base 16) → 10

1_000_000  # underscores for readability
```

### Float (float)

```python
3.14
-0.5
1.0
1e3      # 1000.0
1e-3     # 0.001
```

### Boolean (bool)

must have capitalized letter. if lowercased will be variable.

```python
True
False
```

### None (NoneType)

always compare using is operator and not `=`

```python
None
x = None

print(x is None)     # correct
print(x == None)     # avoid
```

### list

```python
[]
[1, 2, 3]
[1, "a", True]   # mixed types

print([x * 2 for x in range(5)])
```

### Tuple (tuple)

```python
(1, 2, 3)
1, 2, 3        # also valid (commas define tuple)

(1,)           # single element tuple (IMPORTANT)

print(type((10)))
print(type((10,)))
```

### Dictionary (dict)

```python
{}
{"a": 1, "b": 2}        # always keys will have `"`
{1: "x", 2: "y"}        # unless numbers 


print({"a": 5, "a": 20})

```

### Set

unique values. 
mutable.

```python
print({1, 2, 3, 4,3,2})
set()      # empty set (IMPORTANT)
set([1,2,3,4])

print(type({}))         # this will create a dict not a set. 
```
