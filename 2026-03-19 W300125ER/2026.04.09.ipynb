### function is a first class citizen
a function in python is first of all an object that can be passed using variables and parameters

```python
def add(a, b):
    return a + b

print(add)
```

### Default Values
python's default values are evaluated only once on the first call

```python
def add_name(name, list=[]):
    list.append(name)
    print(list)
    
add_name("daniel")
```

```python
add_name("shalom")
```

If argument is needed to be re-evaluated with every function call so use the following pattern


```python
def add_name(name, list=None):
    if list is None:
        list = []       
    list.append(name)
    print(list)
    
add_name("daniel")
```

```python
add_name("shalom")
```

### Positional Arguments

add `, /,` for before it to be only by position after by either positional or keyword (which is the default)

add `, *,` for after it to be only by keyword before by either positional or keyword (which is the default)

```python
def fn1(a,b,c): 
    print(a,b,c)
    
fn1(1,2,3)
fn1(c=3, b=2, a=1)
fn1(c=3, b=2)
```

```python
def fn1(a,b,c, /): 
    print(a,b,c)
    
fn1(1,2,3)
fn1(c=3, b=2, a=1)
```

```python
def fn1(a,b,c, /, d, e): 
    print(a,b,c, d, e)
    
fn1(1,2,3, 4,5)
fn1(1,2,3, e=5, d=4)
```

```python
## positional-only / either positional or keyword (default) * keyword-only

def fn1(a,b,c,/, f, h, *, d, e): 
    print(a,b,c, d, e, f, h)
    
# fn1(1,2,3, 4,5)  

fn1(1,2,3, e=5, d=4, f=10, h=20)
fn1(1,2,3,10,20, e=5, d=4)
fn1(a=1,b=2,c=3, e=5, d=4)
```

```python
def fn1(a,b,c,*, /, d, e): 
    print(a,b,c, d, e)
    
    
fn1(1,2,3, e=5, d=4)
```

### Scopes

```python
x = 10

def f():
    # print("before x=5", x)   # ❌ UnboundLocalError
    x = 5
    print("after x=5", x)
    
f()
print("after f()", x)
```

```python
x = 10
    
def f():
    global x
    # print("before x=5", x)   # ❌ UnboundLocalError
    x = 5
    print("after x=5", x)
    
f()
print("after f()", x)
```

```python
font_size = 15
font_type = "arial"
text = ""


def change_font_size(new_size=font_size):
    """
    documentation
    """
    global font_size
    font_size = new_size


change_font_size(20)
print(font_size)

change_font_size() # the default is still 15 because python evaluates it only once
print(font_size)
    
```
