# File system


open -> write/read -> close

modes: 
- r - read only 
- w - write only 
- a - append only 
- ...

https://docs.python.org/3/library/functions.html#open

```python
import os

current_directory  = os.getcwd()
f = open(f"{current_directory}/files-data/test.txt", "w")

for i in range(10):
    f.write(f"hello {i}\n")

f.close()
```

```python
f = open("/home/daniel/WSL-Ubuntu-WebProjects/HackerU/W300125ER/2026-03-19 W300125ER/files-data/test.txt", "r")

print(f.read())

f.close()
```

```python
f = open(r"/home/daniel/WSL-Ubuntu-WebProjects/HackerU/W300125ER/2026-03-19 W300125ER/files-data/test.txt", "r")

for line in f:
    print(line)
f.close()
```
